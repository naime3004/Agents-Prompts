```python
from crewai import Agent, Task, Crew, Process, tools
```

```python
from re import VERBOSE

## VersÃ£o otimizado pelo ChatGPT

agente_especialista_didatico = Agent(
    role='Especialista em DidÃ¡tica para explicar o tema "{topic}"',
    goal='Explicar conceitos tÃ©cnicos de forma clara, simples e acessÃ­vel, mesmo para quem nÃ£o tem conhecimento prÃ©vio',
    backstory=(
        'VocÃª Ã© um educador apaixonado por ensinar. Possui amplo conhecimento em TI e domina tÃ©cnicas de '
        'comunicaÃ§Ã£o didÃ¡tica. Seu objetivo Ã© traduzir conteÃºdos complexos em explicaÃ§Ãµes fÃ¡ceis, utilizando '
        'linguagem acessÃ­vel, analogias e exemplos prÃ¡ticos.'
    ),
    verbose=True

)

tarefa_criar_resumo_didatico = Task(
    description=(
        'Crie um resumo didÃ¡tico sobre o tema "{topic}". Use linguagem simples, evite jargÃµes tÃ©cnicos sempre que possÃ­vel, '
        'e explique o conteÃºdo de forma progressiva (do bÃ¡sico ao mais avanÃ§ado, se necessÃ¡rio).'
    ),
    expected_output=(
        'Texto em formato Markdown, com tÃ³picos claros, exemplos prÃ¡ticos e, se Ãºtil, comparaÃ§Ãµes ou analogias. '
        'Use listas, destaques e subtÃ­tulos para facilitar a leitura.'
    ),
    output_file='resumo_didatico.md',
    agent=agente_especialista_didatico
)
```

```python
agente_exercicios_didaticos = Agent(
    role='Especialista em Aprendizagem Ativa para o tema "{topic}"',
    goal='Criar exercÃ­cios prÃ¡ticos e reflexivos com base em conteÃºdos tÃ©cnicos, para facilitar a fixaÃ§Ã£o do aprendizado',
    backstory=(
        'VocÃª Ã© um instrutor experiente em Tecnologia da InformaÃ§Ã£o, com foco em aprendizagem ativa. '
        'Sua missÃ£o Ã© transformar conteÃºdos teÃ³ricos em atividades prÃ¡ticas que estimulem o raciocÃ­nio, aplicaÃ§Ã£o e retenÃ§Ã£o. '
        'VocÃª entende que bons exercÃ­cios ajudam o aluno a internalizar o conhecimento de forma duradoura.'
    ),
    verbose=True
)

tarefa_criar_exercicios_didaticos = Task(
    description=(
        'Com base no resumo didÃ¡tico sobre "{topic}", crie uma lista de exercÃ­cios prÃ¡ticos '
        'e reflexivos que ajudem o aluno a fixar os principais conceitos. Os exercÃ­cios devem envolver aplicaÃ§Ã£o prÃ¡tica, identificaÃ§Ã£o de erros, '
        'anÃ¡lise de cenÃ¡rios e perguntas abertas.'
    ),
    expected_output=(
        'Arquivo em Markdown contendo de 5 a 10 exercÃ­cios, numerados e organizados por nÃ­vel de dificuldade (bÃ¡sico, intermediÃ¡rio, avanÃ§ado). '
        'Incluir instruÃ§Ãµes claras e, quando apropriado, pequenos trechos de cÃ³digo ou situaÃ§Ãµes do dia a dia.'
    ),
    output_file='exercicios_didaticos.md',
    agent=agente_exercicios_didaticos
)
```

```python
equipe = Crew(
    agents=[agente_especialista_didatico, agente_exercicios_didaticos ],
    tasks=[tarefa_criar_resumo_didatico, tarefa_criar_exercicios_didaticos ],
    process=Process.sequential
)

entrada_usuario = {
    'topic': 'json'
}

resultado = equipe.kickoff(inputs=entrada_usuario)
print(resultado)
```
