```python
from crewai import Agent, Task, Crew, Process, tools
```

```python
from re import VERBOSE

## Versão otimizado pelo ChatGPT

agente_especialista_didatico = Agent(
    role='Especialista em Didática para explicar o tema "{topic}"',
    goal='Explicar conceitos técnicos de forma clara, simples e acessível, mesmo para quem não tem conhecimento prévio',
    backstory=(
        'Você é um educador apaixonado por ensinar. Possui amplo conhecimento em TI e domina técnicas de '
        'comunicação didática. Seu objetivo é traduzir conteúdos complexos em explicações fáceis, utilizando '
        'linguagem acessível, analogias e exemplos práticos.'
    ),
    verbose=True

)

tarefa_criar_resumo_didatico = Task(
    description=(
        'Crie um resumo didático sobre o tema "{topic}". Use linguagem simples, evite jargões técnicos sempre que possível, '
        'e explique o conteúdo de forma progressiva (do básico ao mais avançado, se necessário).'
    ),
    expected_output=(
        'Texto em formato Markdown, com tópicos claros, exemplos práticos e, se útil, comparações ou analogias. '
        'Use listas, destaques e subtítulos para facilitar a leitura.'
    ),
    output_file='resumo_didatico.md',
    agent=agente_especialista_didatico
)
```

```python
import os
from google.colab import userdata

os.environ['OPENAI_API_KEY'] = userdata.get('OPENAI_API_KEY')
```

```python
agente_exercicios_didaticos = Agent(
    role='Especialista em Aprendizagem Ativa para o tema "{topic}"',
    goal='Criar exercícios práticos e reflexivos com base em conteúdos técnicos, para facilitar a fixação do aprendizado',
    backstory=(
        'Você é um instrutor experiente em Tecnologia da Informação, com foco em aprendizagem ativa. '
        'Sua missão é transformar conteúdos teóricos em atividades práticas que estimulem o raciocínio, aplicação e retenção. '
        'Você entende que bons exercícios ajudam o aluno a internalizar o conhecimento de forma duradoura.'
    ),
    verbose=True
)

tarefa_criar_exercicios_didaticos = Task(
    description=(
        'Com base no resumo didático sobre "{topic}", crie uma lista de exercícios práticos '
        'e reflexivos que ajudem o aluno a fixar os principais conceitos. Os exercícios devem envolver aplicação prática, identificação de erros, '
        'análise de cenários e perguntas abertas.'
    ),
    expected_output=(
        'Arquivo em Markdown contendo de 5 a 10 exercícios, numerados e organizados por nível de dificuldade (básico, intermediário, avançado). '
        'Incluir instruções claras e, quando apropriado, pequenos trechos de código ou situações do dia a dia.'
    ),
    output_file='exercicios_didaticos.md',
    agent=agente_exercicios_didaticos
)
```

```python
equipe = Crew(
    agents=[agente_especialista_didatico, agente_exercicios_didaticos ],
    tasks=[tarefa_criar_resumo_didatico, tarefa_criar_exercicios_didaticos ],
    process=Process.sequential
)

entrada_usuario = {
    'topic': 'json'
}

resultado = equipe.kickoff(inputs=entrada_usuario)
print(resultado)
```
